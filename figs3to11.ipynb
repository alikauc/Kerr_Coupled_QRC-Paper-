#### Binned analysis function

```julia
# This function will bin and plot our desired data
function binned(ent,nmrse,xlabel,ylabel,k,κ,κϕ,type,f)
    # Define bin edges for entanglement (e.g., 0.0-0.2, 0.2-0.4, etc.)
    bin_edges = 0:0.05:0.3
    # Initialize an array to store the average NRMSE for each bin
    avg_nrmse_per_bin = []
    var_nrmse_per_bin = []
    confidence_intervals=[]
    # Loop through each bin to calculate the average NRMSE
    for i in 1:(length(bin_edges) - 1)
        # Get the indices of entanglement values that fall within the current bin
        bin_mask = (ent .>= bin_edges[i]) .& (ent .< bin_edges[i+1])
        
        # Filter the NRMSE values corresponding to the current bin
        nrmse_in_bin = nmrse[bin_mask]
        
        # Calculate the average NRMSE in this bin
        if length(nrmse_in_bin) > 0
            if type=="mean"
                push!(avg_nrmse_per_bin, mean(nrmse_in_bin))
            elseif type=="max"
                push!(avg_nrmse_per_bin, maximum(nrmse_in_bin))
            elseif type=="min"
                push!(avg_nrmse_per_bin, minimum(nrmse_in_bin))
            end
            push!(var_nrmse_per_bin, std(nrmse_in_bin))
            # Calculate 95% confidence interval using standard error
            n = length(nrmse_in_bin)
            std_err = sqrt(var(nrmse_in_bin) / n)
            ci_95 = 1.96 * std_err  # 1.96 for 95% confidence interval
            push!(confidence_intervals, ci_95)
        else
            push!(avg_nrmse_per_bin, NaN)  # If no data in the bin, push NaN
        end
    end
    # Plot the average NRMSE vs. entanglement bin ranges
    bin_centers = (bin_edges[1:end-1] .+ bin_edges[2:end]) ./ 2
    bin_widths = (bin_edges[2:end] .- bin_edges[1:end-1]) ./ 2  # Half-width of each bin for error bars
    if !isdefined(Main, :pb1)||isnothing(pb1)
        mycolor=:auto
        theme(:default, legendfontsize=20, guidefontsize=20,tickfontsize=20,titlefont=font(20)
        ,grid=:false)
        global pb1=plt.plot(bin_centers, avg_nrmse_per_bin,
        xerror=bin_widths,yerror=var_nrmse_per_bin,
        color=mycolor,
        linecolor=mycolor,
        markercolor=mycolor,
        markerstrokecolor=mycolor,
        errorbarcolor=mycolor,  # Some versions of GR may support this
        errorbar_linewidth=0.5,
        linewidth=1.9,
        markershape=:auto,
        markersize=5,
        xlabel=xlabel, ylabel=ylabel,title="κ=$κ, κϕ=$κϕ, K=$k", label="f=$f",size=(600,750), marker=:auto)
        pb2=pb1
    else
        mycolor=:auto
        plt.plot!(pb1,bin_centers, avg_nrmse_per_bin,
        xerror=bin_widths,yerror=var_nrmse_per_bin,
        color=mycolor,
        linecolor=mycolor,
        markercolor=mycolor,
        markerstrokecolor=mycolor,
        errorbarcolor=mycolor,
        errorbar_linewidth=0.5,
        linewidth=1.9,
        markershape=:auto,
        markersize=5,
        xlabel=xlabel, ylabel=ylabel,title="κ=$κ, κϕ=$κϕ, K=$k", label="f=$f", marker=:auto,legend=:outertopright)
    end
end
```
